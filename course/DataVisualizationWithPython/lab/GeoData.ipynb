```python
df_incidents = pd.read_csv('criminal_data.csv')

print('Dataset downloaded and read into a pandas dataframe!')
```

```python
# get the first 100 crimes in the df_incidents dataframe
limit = 300
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42
```

```python
sanfran_map = folium.Map(location=[latitude,longitude], zoom_start=12)
```

```python
incidents = folium.map.FeatureGroup()

for lat,long in zip(df_incidents.Y,df_incidents.X):
    incidents.add_child(
        folium.features.CircleMarker(
        [lat,long],
        radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.8
        )
    )
sanfran_map.add_child(incidents)
```

```python
san_map = folium.Map(location=[latitude,longitude], zoom_start=12)
incidents = folium.features.FeatureGroup()

for lat,long in zip(df_incidents.Y,df_incidents.X):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat,long],
            radius=5,
            color='red',
            fill=True,
            fill_color='yesllow',
            fill_opacity=0.8
        )
    )
san_map.add_child(incidents)

latitudes = list(df_incidents.Y)
longitudes = list(df_incidents.X)
labels = list(df_incidents.Category)


for lat,lon,label in zip(latitudes,longitudes,labels):
    folium.Marker([lat,lon],popup=label).add_to(san_map)
    
san_map.add_child(incidents)
```

```python
from folium import plugins

sanfran_map = folium.Map(location=[latitude,longitude],zoom_start=12)
incidents = plugins.MarkerCluster().add_to(sanfran_map)

for lat,long,label in zip(df_incidents.Y,df_incidents.X,df_incidents.Category):
    folium.Marker(
        location=[lat,long],
        icon=None,
        popup=label
    ).add_to(incidents)
sanfran_map
```
